```python
import pandas as pd
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium
from geopy.geocoders import Nominatim

df = pd.read_csv('Geospatial_Coordinates.csv') 
d = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
```

```python
d = d[d.Borough != 'Not assigned']
d['Neighborhood'].loc[d['Neighborhood'] == 'Not assigned'] = d['Borough'].loc[d['Neighborhood'] == 'Not assigned']
d = d.reset_index()
d.drop(['index'], axis=1)
```

```python
df_join = pd.merge(d, df, on='Postal Code', how='outer')
df_join.drop(['index'], axis=1)
```

#### Latitude and Longitude of Toronto

```python
latitude = 43.6532
longitude = -79.3832
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

#### Create a map of Toronto with neighborhoods superimposed on top.

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_join['Latitude'], df_join['Longitude'], df_join['Borough'], df_join['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
